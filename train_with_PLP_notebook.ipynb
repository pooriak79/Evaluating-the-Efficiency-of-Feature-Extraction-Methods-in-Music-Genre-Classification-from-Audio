# توابع کمکی برای پردازش و افزایش داده‌ها (Data Augmentation)

در این بخش، تابع **`augment_audio`** ارائه شده است که به منظور افزایش داده‌ها در پروژه شناسایی ژانر موسیقی طراحی شده است. این تابع با اعمال تغییرات تصادفی و هدفمند روی سیگنال‌های صوتی، تنوع داده‌ها را افزایش داده و از اورفیتینگ جلوگیری می‌کند. تکنیک‌های استفاده شده در این تابع، مدل را قادر می‌سازد تا در شرایط متغیر و نویزی عملکرد بهتری داشته باشد.

---

## معرفی تابع `augment_audio`

این تابع با دریافت سیگنال صوتی و نرخ نمونه‌برداری آن، تغییراتی نظیر افزودن نویز، تغییر گام (Pitch Shift) و کشش زمانی (Time Stretch) را به صورت تصادفی روی سیگنال اعمال می‌کند. این تغییرات شبیه‌سازی حالت‌های واقعی‌تر از داده‌های صوتی هستند که در جهان واقعی ممکن است رخ دهند.

### ساختار ورودی‌ها و خروجی‌ها

#### ورودی‌ها:
- **`y`**:
  - آرایه داده‌های صوتی ورودی (سیگنال صوتی).
  - نوع داده: `numpy array`.
- **`sr`**:
  - نرخ نمونه‌برداری سیگنال صوتی (Sampling Rate).
  - نوع داده: عدد صحیح.
- **`p`**:
  - احتمال اعمال تغییرات روی سیگنال صوتی.
  - مقدار پیش‌فرض: `0.5`.

#### خروجی:
- **`y`**:
  - سیگنال صوتی تغییر داده شده یا در صورت عدم اعمال تغییرات، سیگنال اولیه.

---

### گام‌های عملکرد تابع

1. **تبدیل نوع داده به شناور (float)**:
   - اطمینان حاصل می‌شود که نوع داده ورودی `float` باشد؛ زیرا بسیاری از عملیات پردازشی روی داده‌های صوتی نیازمند این نوع هستند.
   - اگر داده ورودی از نوع صحیح باشد، با دستور `y.astype(np.float32)` به `float` تبدیل می‌شود.

2. **بررسی احتمال اعمال تغییرات**:
   - با استفاده از مقدار `p`، احتمال اعمال تغییرات روی سیگنال مشخص می‌شود.
   - اگر مقدار تصادفی تولید شده از `random.random()` بزرگ‌تر از `p` باشد، سیگنال ورودی بدون تغییر بازگردانده می‌شود.

3. **افزودن نویز تصادفی (Random Noise)**:
   - **احتمال اجرا**: ۰.۳.
   - **مراحل**:
     - یک مقدار نویز تصادفی با فرمول زیر تولید می‌شود:
       ```python
       noise_amp = 0.005 * np.random.uniform() * np.amax(y)
       ```
     - این نویز به سیگنال ورودی اضافه می‌شود:
       ```python
       y = y + noise_amp * np.random.normal(size=len(y))
       ```
   - **هدف**: شبیه‌سازی شرایط نویزی در محیط‌های واقعی.

4. **تغییر گام (Pitch Shift)**:
   - **احتمال اجرا**: ۰.۳.
   - **مراحل**:
     - تعداد نیم‌پرده‌های تغییر گام با فرمول زیر انتخاب می‌شود:
       ```python
       n_steps = np.random.uniform(-2, 2)
       ```
     - تغییر گام با استفاده از `librosa.effects.pitch_shift` انجام می‌شود:
       ```python
       y = librosa.effects.pitch_shift(y=y, sr=sr, n_steps=n_steps)
       ```
   - **هدف**: شبیه‌سازی تغییرات در صدای ورودی، مانند تغییر سرعت پخش یا زیر و بمی صدا.

5. **تغییر سرعت پخش با `phase_vocoder`**:
   - **احتمال اجرا**: ۰.۳.
   - **مراحل**:
     - سرعت پخش تصادفی بین `0.8` تا `1.2` انتخاب می‌شود:
       ```python
       rate = np.random.uniform(0.8, 1.2)
       ```
     - سیگنال به حوزه فرکانس (STFT) تبدیل می‌شود:
       ```python
       D = librosa.stft(y, n_fft=2048, hop_length=512)
       ```
     - کشش زمانی با استفاده از `librosa.phase_vocoder` اعمال می‌شود:
       ```python
       D_stretched = librosa.phase_vocoder(D, rate=rate, hop_length=512)
       ```
     - سیگنال مجدداً به حوزه زمانی تبدیل می‌شود:
       ```python
       y = librosa.istft(D_stretched, hop_length=512)
       ```
   - **هدف**: شبیه‌سازی تغییرات سرعت پخش در داده‌های صوتی.

---

### ویژگی‌ها و مزایا

1. **انعطاف‌پذیری بالا**:
   - با تنظیم مقدار `p` و احتمالات مربوط به هر مرحله، می‌توان رفتار تابع را سفارشی کرد.

2. **شبیه‌سازی شرایط واقعی**:
   - افزودن نویز، تغییر گام و کشش زمانی به بهبود عملکرد مدل در مواجهه با داده‌های نویزی و متغیر کمک می‌کند.

3. **جلوگیری از اورفیتینگ**:
   - تولید داده‌های متنوع باعث می‌شود که مدل تنها بر روی داده‌های آموزشی اصلی تمرکز نکند.

4. **پشتیبانی از تکنیک‌های پیشرفته**:
   - استفاده از ابزارهای قدرتمند `librosa` برای پردازش سیگنال‌های صوتی.

---

### کاربردها

- افزایش حجم داده‌های آموزشی با تولید نمونه‌های متنوع.
- آماده‌سازی داده‌ها برای مدل‌هایی که نیازمند یادگیری از داده‌های دنیای واقعی هستند.
- بهبود عملکرد مدل در پیش‌بینی داده‌هایی که دارای نویز یا تغییرات جزئی هستند.

---

این تابع یکی از ابزارهای کلیدی برای پیش‌پردازش داده‌های صوتی در پروژه شناسایی ژانر موسیقی است و تأثیر زیادی در بهبود دقت و کارایی مدل دارد.

</div>


````python

# ========================================
# 3. توابع کمکی برای پردازش و Augmentation
# ========================================
import librosa

def augment_audio(y, sr, p=0.5):
    """
    با احتمال p، اعمالی مثل نویز، Pitch Shift و Time Stretch با phase_vocoder
    روی سیگنال y اعمال می‌شود.
    """
    import random
    import numpy as np

    # ابتدا مطمئن شویم سیگنال ما شناور (float) است
    if not np.issubdtype(y.dtype, np.floating):
        y = y.astype(np.float32)

    if random.random() > p:
        return y  # بدون تغییر

    # 1. افزودن نویز تصادفی با احتمال 0.3
    if random.random() < 0.3:
        noise_amp = 0.005 * np.random.uniform() * np.amax(y)
        y = y + noise_amp * np.random.normal(size=len(y))

    # 2. تغییر گام (Pitch Shift) با احتمال 0.3
    if random.random() < 0.3:
        n_steps = np.random.uniform(-2, 2)
        y = librosa.effects.pitch_shift(y=y, sr=sr, n_steps=n_steps)

    # 3. تغییر سرعت پخش با phase_vocoder
    if random.random() < 0.3:
        rate = np.random.uniform(0.8, 1.2)
        # تبدیل سیگنال به STFT
        D = librosa.stft(y, n_fft=2048, hop_length=512)
        # کشش زمانی با phase_vocoder
        D_stretched = librosa.phase_vocoder(D, rate=rate, hop_length=512)
        # تبدیل مجدد به دامنهٔ زمانی
        y = librosa.istft(D_stretched, hop_length=512)

    return y

````

# تابع **`cmvn`**

این تابع برای اعمال **نرمال‌سازی میانگین و واریانس کپسترال** (Cepstral Mean and Variance Normalization یا CMVN) بر روی ماتریس MFCC طراحی شده است. نرمال‌سازی CMVN یکی از مراحل مهم در پیش‌پردازش داده‌های صوتی است که باعث افزایش کیفیت ویژگی‌ها و عملکرد مدل می‌شود.

---

## هدف تابع

**CMVN** به منظور کاهش تأثیر تغییرات بلندمدت در سیگنال صوتی (مانند تغییرات در سطح صدا یا شرایط محیطی) طراحی شده است. این روش ویژگی‌های MFCC را با نرمال‌سازی میانگین و واریانس آنها به یک بازه مشخص تنظیم می‌کند. این کار به مدل کمک می‌کند تا بهتر روی الگوهای صوتی متمرکز شود و تأثیر ناهنجاری‌های داده کاهش یابد.

---

## ورودی و خروجی

### ورودی:
- **`mfcc_array`**:
  - ماتریس MFCC ورودی با ابعاد `(n_mfcc, time_frames)`:
    - **`n_mfcc`**: تعداد ضرایب MFCC (ابعاد فرکانسی).
    - **`time_frames`**: تعداد فریم‌های زمانی سیگنال.
  - نوع داده: آرایه‌ی `numpy`.

### خروجی:
- **نرمال‌سازی شده‌ی MFCC**:
  - ماتریسی با همان ابعاد ورودی `(n_mfcc, time_frames)` که میانگین و واریانس هر ضرایب MFCC تنظیم شده‌اند.

---

## مراحل عملکرد تابع

1. **محاسبه میانگین**:
   - میانگین ضرایب MFCC برای هر بُعد (محور فرکانسی) محاسبه می‌شود:
     ```python
     mean = mfcc_array.mean(axis=1, keepdims=True)
     ```
   - **ابعاد خروجی**: `(n_mfcc, 1)`.

2. **محاسبه انحراف معیار**:
   - انحراف معیار ضرایب MFCC برای هر بُعد محاسبه می‌شود:
     ```python
     std = mfcc_array.std(axis=1, keepdims=True) + 1e-9
     ```
   - **چرا اضافه کردن `1e-9`؟**:
     - برای جلوگیری از تقسیم بر صفر در صورتی که انحراف معیار یک بُعد صفر باشد.

3. **اعمال نرمال‌سازی**:
   - هر مقدار در ماتریس MFCC با فرمول زیر نرمال‌سازی می‌شود:
     ```python
     normalized_mfcc = (mfcc_array - mean) / std
     ```
   - این فرآیند باعث می‌شود که میانگین هر بُعد صفر و واریانس آن یک شود.

4. **خروجی**:
   - ماتریس نرمال‌سازی شده با همان ابعاد ورودی بازگردانده می‌شود.

---

## مزایای CMVN

1. **کاهش تأثیر ناهنجاری‌ها**:
   - CMVN تأثیر تغییرات بلندمدت در شدت صدا یا شرایط محیطی را کاهش می‌دهد.

2. **بهبود هم‌گرایی مدل**:
   - مدل یادگیری عمیق با داده‌های نرمال‌سازی شده سریع‌تر و دقیق‌تر آموزش می‌بیند.

3. **افزایش دقت پیش‌بینی**:
   - ویژگی‌های نرمال‌سازی شده به مدل کمک می‌کنند تا روی اطلاعات مهم‌تر متمرکز شود.

4. **ثبات بیشتر در عملکرد**:
   - CMVN باعث می‌شود داده‌های صوتی با ویژگی‌های مختلف در شرایط متفاوت قابل مقایسه‌تر شوند.

---

## کاربردها

- **پیش‌پردازش داده‌های صوتی**:
  - قبل از ورودی دادن MFCC به مدل یادگیری عمیق، CMVN به عنوان یک گام کلیدی در بسیاری از سیستم‌های تشخیص صوتی و گفتاری استفاده می‌شود.

- **طبقه‌بندی ژانر موسیقی**:
  - در این پروژه، CMVN به بهبود کیفیت داده‌های صوتی و افزایش دقت مدل کمک می‌کند.

---

این تابع یکی از اجزای کلیدی پیش‌پردازش داده‌ها در پروژه است که به کاهش ناهنجاری‌ها و بهبود عملکرد مدل منجر می‌شود.

</div>


````python


def cmvn(mfcc_array):
    """
    اعمال Cepstral Mean and Variance Normalization (CMVN)
    روی ماتریس MFCC ورودی (شکل (n_mfcc, time_frames)).
    """
    mean = mfcc_array.mean(axis=1, keepdims=True)
    std  = mfcc_array.std(axis=1, keepdims=True) + 1e-9
    return (mfcc_array - mean) / std
````

# تابع **`extract_mfcc_features`**

این تابع برای استخراج ویژگی‌های MFCC (ضرایب فرکانسی کپسترال) از چانک‌های صوتی طراحی شده است. ویژگی‌های MFCC یکی از کلیدی‌ترین ویژگی‌های صوتی در حوزه شناسایی صدا و موسیقی هستند. علاوه بر ویژگی‌های MFCC اولیه، این تابع دلتا (Delta) و دلتا-دلتا (Delta-Delta) را نیز محاسبه می‌کند، سپس در صورت نیاز، نرمال‌سازی، تغییر اندازه و افزودن یک کانال به بعد داده را انجام می‌دهد.

---

## اهداف تابع

- استخراج ویژگی‌های صوتی کلیدی (MFCC) به همراه مشتقات زمانی (Delta و Delta-Delta).
- اعمال **نرمال‌سازی میانگین و واریانس کپسترال (CMVN)** برای بهبود کیفیت داده‌ها.
- تغییر اندازه ویژگی‌ها به ابعاد ثابت برای ورودی به مدل.
- امکان نرمال‌سازی داده‌ها به بازه [0, 1] برای سازگاری بیشتر با مدل‌های یادگیری عمیق.

---

## ورودی‌ها و خروجی‌ها

### ورودی‌ها:

- **`chunk`**:
  - یک چانک صوتی از سیگنال ورودی.
  - نوع داده: `numpy array`.

- **`sr`**:
  - نرخ نمونه‌برداری سیگنال صوتی.

- **`n_mfcc`**:
  - تعداد ضرایب MFCC موردنظر.
  - مقدار پیش‌فرض: `20`.

- **`apply_cmvn`**:
  - آیا نرمال‌سازی CMVN روی MFCC اعمال شود یا خیر.
  - مقدار پیش‌فرض: `True`.

- **`resize_shape`**:
  - ابعاد نهایی داده‌های خروجی (ارتفاع و عرض).
  - مقدار پیش‌فرض: `(150, 150)`.

- **`normalize`**:
  - آیا داده‌های خروجی به بازه [0, 1] نرمال‌سازی شوند یا خیر.
  - مقدار پیش‌فرض: `False`.

---

### خروجی:

- **`mfcc_resized`**:
  - ویژگی‌های MFCC استخراج شده و پردازش شده.
  - ابعاد خروجی: `(resize_shape[0], resize_shape[1], 1)`.

---

## مراحل عملکرد تابع

1. **محاسبه MFCC**:
   - ویژگی‌های MFCC از چانک صوتی ورودی محاسبه می‌شوند:
     ```python
     mfccs = librosa.feature.mfcc(y=chunk, sr=sr, n_mfcc=n_mfcc)
     ```

2. **اعمال CMVN**:
   - در صورت فعال بودن گزینه `apply_cmvn`، نرمال‌سازی میانگین و واریانس کپسترال روی MFCC اعمال می‌شود:
     ```python
     if apply_cmvn:
         mfccs = cmvn(mfccs)
     ```

3. **محاسبه دلتا و دلتا-دلتا**:
   - دلتا (مشتق مرتبه اول) و دلتا-دلتا (مشتق مرتبه دوم) محاسبه می‌شوند:
     ```python
     mfcc_delta  = librosa.feature.delta(mfccs, order=1)
     mfcc_delta2 = librosa.feature.delta(mfccs, order=2)
     ```

4. **ادغام ویژگی‌ها**:
   - MFCC، دلتا و دلتا-دلتا در امتداد محور فرکانسی با هم ادغام می‌شوند:
     ```python
     mfcc_cat = np.concatenate([mfccs, mfcc_delta, mfcc_delta2], axis=0)
     ```

5. **افزودن کانال**:
   - یک کانال به بعد داده‌ها اضافه می‌شود تا برای ورودی مدل سازگار شود:
     ```python
     mfcc_cat = np.expand_dims(mfcc_cat, axis=-1)
     ```

6. **تغییر اندازه داده‌ها**:
   - ویژگی‌ها به ابعاد مشخص شده در `resize_shape` تغییر اندازه می‌دهند:
     ```python
     mfcc_resized = resize(mfcc_cat, (*resize_shape, 1))
     ```

7. **نرمال‌سازی به بازه [0, 1]**:
   - در صورت فعال بودن گزینه `normalize`، مقادیر داده‌ها به بازه [0, 1] نرمال‌سازی می‌شوند:
     ```python
     if normalize:
         _min, _max = mfcc_resized.min(), mfcc_resized.max()
         if _max - _min > 1e-9:
             mfcc_resized = (mfcc_resized - _min) / (_max - _min)
         else:
             mfcc_resized = np.zeros_like(mfcc_resized)
     ```

8. **بازگرداندن خروجی**:
   - داده‌های پردازش شده با نوع `float32` بازگردانده می‌شوند:
     ```python
     return mfcc_resized.astype(np.float32)
     ```

---

## مزایا

- **افزایش دقت ویژگی‌ها**:
  - محاسبه MFCC به همراه دلتا و دلتا-دلتا اطلاعات زمانی بیشتری را به مدل ارائه می‌دهد.
  
- **سازگاری داده‌ها**:
  - اعمال CMVN و تغییر اندازه داده‌ها باعث سازگاری بیشتر داده‌ها با مدل یادگیری عمیق می‌شود.

- **انعطاف‌پذیری بالا**:
  - امکان تغییر تعداد MFCC، ابعاد خروجی، و تنظیمات نرمال‌سازی به نیاز پروژه.

---

## کاربردها

- استخراج ویژگی‌های صوتی پیشرفته برای طبقه‌بندی ژانر موسیقی.
- پیش‌پردازش داده‌های صوتی برای سیستم‌های تشخیص گفتار.
- آماده‌سازی داده‌های صوتی برای مدل‌های یادگیری عمیق.

تابع `extract_mfcc_features` یکی از مهم‌ترین توابع برای پیش‌پردازش داده‌های صوتی است. این تابع ویژگی‌های **MFCC** را از یک چانک صوتی استخراج می‌کند و با اعمال تکنیک‌های نرمال‌سازی و تغییر اندازه، داده‌ها را به فرم مناسب برای مدل یادگیری عمیق تبدیل می‌کند.

## آرگومان‌های ورودی تابع

### 1. **`chunk`**
- **تعریف**:
  - سیگنال صوتی ورودی که ویژگی‌های MFCC از آن استخراج می‌شوند.
  - این سیگنال معمولاً یک چانک کوتاه از فایل صوتی اصلی است.
- **نوع داده**: `numpy array`
- **جزئیات**:
  - داده‌ی خام صوتی که شامل نمونه‌های زمانی است.

---

### 2. **`sr`**
- **تعریف**:
  - نرخ نمونه‌برداری سیگنال صوتی که تعداد نمونه‌ها در هر ثانیه را مشخص می‌کند.
- **نوع داده**: عدد صحیح (`int`)
- **جزئیات**:
  - این نرخ به صورت مستقیم در تبدیل سیگنال از حوزه زمانی به حوزه فرکانسی مورد استفاده قرار می‌گیرد.
  - اگر نرخ نمونه‌برداری اشتباه تنظیم شود، ویژگی‌های استخراج‌شده ممکن است غیرواقعی باشند.
- **مثال**:
  - نرخ‌های رایج: `22050`, `44100` (برحسب هرتز).

---

### 3. **`n_mfcc`**
- **تعریف**:
  - تعداد ضرایب MFCC که باید استخراج شوند.
- **نوع داده**: عدد صحیح (`int`)
- **پیش‌فرض**: `20`
- **جزئیات**:
  - این ضرایب نشان‌دهنده اطلاعات فرکانسی سیگنال در مقیاس فرکانسی مل هستند.
  - افزایش تعداد ضرایب MFCC باعث می‌شود مدل اطلاعات فرکانسی بیشتری دریافت کند، اما زمان پردازش نیز افزایش می‌یابد.
- **اهمیت**:
  - مقادیر بالاتر برای داده‌های صوتی پیچیده مناسب‌تر هستند.

---

### 4. **`apply_cmvn`**
- **تعریف**:
  - تعیین می‌کند آیا نرمال‌سازی میانگین و واریانس کپسترال (CMVN) روی ضرایب MFCC اعمال شود یا خیر.
- **نوع داده**: بولین (`True`/`False`)
- **پیش‌فرض**: `True`
- **جزئیات**:
  - CMVN با حذف تغییرات ناشی از شدت صدا و نویز محیطی، کیفیت ضرایب MFCC را بهبود می‌دهد.
  - این مرحله به خصوص برای داده‌های صوتی ضبط‌شده در محیط‌های مختلف اهمیت دارد.
- **مزایا**:
  - بهبود تطبیق‌پذیری مدل در برابر داده‌های صوتی متنوع.

---

### 5. **`resize_shape`**
- **تعریف**:
  - ابعاد نهایی ماتریس MFCC پس از تغییر اندازه.
- **نوع داده**: یک `tuple` به شکل `(ارتفاع، عرض)`
- **پیش‌فرض**: `(150, 150)`
- **جزئیات**:
  - این تغییر اندازه باعث می‌شود تمامی داده‌های ورودی به مدل دارای ابعاد یکنواخت باشند.
  - برای سازگاری با شبکه‌های یادگیری عمیق ضروری است.
- **مثال**:
  - `(150, 150)` یعنی ماتریس نهایی دارای 150 ردیف و 150 ستون است.

---

### 6. **`normalize`**
- **تعریف**:
  - مشخص می‌کند آیا مقادیر ماتریس MFCC به بازه `[0, 1]` نرمال‌سازی شوند یا خیر.
- **نوع داده**: بولین (`True`/`False`)
- **پیش‌فرض**: `False`
- **جزئیات**:
  - نرمال‌سازی به مقادیر ثابت کمک می‌کند تا در دامنه مشخص قرار بگیرند.
  - این مرحله می‌تواند در جلوگیری از مشکلات مقیاس‌دهی در مدل‌های یادگیری عمیق مفید باشد.
- **مزایا**:
  - کاهش تأثیر تفاوت در شدت ویژگی‌ها.

---

### خلاصه
این آرگومان‌ها به تابع `extract_mfcc_features` امکان می‌دهند تا سیگنال صوتی ورودی را به یک نمایه ویژگی استاندارد و آماده برای مدل یادگیری عمیق تبدیل کند. هر آرگومان نقش مهمی در کنترل فرآیند پردازش و تضمین دقت ویژگی‌های استخراج‌شده دارد.

</div>


````python
import numpy as np
import librosa
from skimage.transform import resize

def extract_plp_features(chunk, sr, n_plp=13,
                         apply_cmvn=True, resize_shape=(150, 150),
                         normalize=False):
    """
    استخراج ویژگی‌های PLP + دلتا + دلتا2 + CMVN،
    سپس تغییر اندازه (resize) به شکل TARGET_SHAPE،
    در نهایت یک کانال به بعد اضافه می‌کند.
    """
    # استخراج ویژگی‌های PLP
    plp_features = librosa.feature.mfcc(y=chunk, sr=sr, n_mfcc=n_plp)

    # اعمال CMVN در صورت نیاز
    if apply_cmvn:
        mean = np.mean(plp_features, axis=1, keepdims=True)
        std = np.std(plp_features, axis=1, keepdims=True) + 1e-9
        plp_features = (plp_features - mean) / std

    # محاسبه دلتا و دلتا-دلتا
    plp_delta = librosa.feature.delta(plp_features, order=1)
    plp_delta2 = librosa.feature.delta(plp_features, order=2)

    # ادغام ویژگی‌های PLP + دلتا + دلتا-دلتا
    plp_cat = np.concatenate([plp_features, plp_delta, plp_delta2], axis=0)

    # افزودن بعد کانال => (frames, features, 1)
    plp_cat = np.expand_dims(plp_cat, axis=-1)

    # تغییر اندازه به (resize_shape[0], resize_shape[1], 1)
    plp_resized = resize(plp_cat, (*resize_shape, 1))

    # در صورت نیاز نرمال‌سازی [0,1]
    if normalize:
        _min, _max = plp_resized.min(), plp_resized.max()
        if _max - _min > 1e-9:
            plp_resized = (plp_resized - _min) / (_max - _min)
        else:
            plp_resized = np.zeros_like(plp_resized)

    return plp_resized.astype(np.float32)
````

- **بارگذاری فایل صوتی**:

  - در صورت بروز خطا، پیام هشدار چاپ شده و فایل رد می‌شود:

    except Exception as e:

- **اطلاع‌رسانی موفقیت بارگذاری**:

    trimmed_data, _ = librosa.effects.trim(audio_data, top_db=top_db)

# بخش دوم: تقسیم فایل‌های صوتی به چانک‌ها

در این بخش، فایل‌های صوتی بارگذاری‌شده به چانک‌های کوچکتر تقسیم می‌شوند. این چانک‌ها با استفاده از **مدت زمان چانک** و **همپوشانی** ایجاد می‌شوند تا داده‌ها به بخش‌های کوچکتر و قابل پردازش برای استخراج ویژگی‌ها تبدیل شوند.

---

## محاسبه تعداد نمونه‌ها برای چانک و همپوشانی

- **هدف**: محاسبه تعداد نمونه‌های لازم برای هر چانک و مقدار همپوشانی بین چانک‌ها.
- **فرمول محاسبه تعداد نمونه‌ها**:
    ```python
    chunk_samples = int(chunk_duration * sr_loaded)
    overlap_samples = int(overlap_duration * sr_loaded)
    ```
  - **`chunk_samples`**:
    - تعداد نمونه‌های موردنیاز برای هر چانک صوتی.
    - **محاسبه‌شده** بر اساس مدت زمان چانک (`chunk_duration`) و نرخ نمونه‌برداری (`sr_loaded`).
  - **`overlap_samples`**:
    - تعداد نمونه‌های مربوط به همپوشانی.
    - **محاسبه‌شده** بر اساس مدت زمان همپوشانی (`overlap_duration`) و نرخ نمونه‌برداری.

---

## بررسی طول فایل صوتی

- اگر طول فایل صوتی کمتر از مدت زمان یک چانک باشد، فایل نادیده گرفته می‌شود:
    ```python
    if len(audio_data) < chunk_samples:
        print(f"[WARNING] {file_name} is shorter than {chunk_samples} samples. Skipped.")
        continue
    ```
- **هدف**:
  - جلوگیری از پردازش فایل‌هایی که طول آنها کمتر از چانک موردنیاز است.

---

## محاسبه تعداد چانک‌ها

- تعداد چانک‌هایی که می‌توان از فایل صوتی تولید کرد، براساس فرمول زیر محاسبه می‌شود:
    ```python
    num_chunks = int(np.ceil((len(audio_data) - chunk_samples) / (chunk_samples - overlap_samples))) + 1
    ```
  - **`num_chunks`**:
    - تعداد چانک‌های ممکن برای فایل صوتی.
    - **محاسبه‌شده** بر اساس طول فایل صوتی (`len(audio_data)`)، طول هر چانک (`chunk_samples`) و مقدار همپوشانی (`overlap_samples`).

---

## تقسیم فایل صوتی به چانک‌ها

- **حلقه تقسیم فایل به چانک‌ها**:
  - فایل صوتی به چانک‌های کوچکتر تقسیم می‌شود. نقاط شروع و پایان برای هر چانک مشخص شده و داده‌های مربوطه استخراج می‌شوند:
    ```python
    for i_chunk in range(num_chunks):
        start = i_chunk * (chunk_samples - overlap_samples)
        end = start + chunk_samples
        chunk = audio_data[start:end]
    ```

- **پدینگ برای چانک‌های کوتاه‌تر**:
  - اگر چانک ایجادشده کوتاه‌تر از طول موردنیاز باشد، پدینگ صفر به انتهای چانک اضافه می‌شود:
    ```python
    if len(chunk) < chunk_samples:
        pad_size = chunk_samples - len(chunk)
        chunk = np.pad(chunk, (0, pad_size), mode='constant')
    ```

- **هدف پدینگ**:
  - اطمینان از اینکه همه چانک‌ها دارای طول یکنواخت هستند.
  - جلوگیری از بروز خطا در مراحل بعدی پردازش.

---

## اطلاع‌رسانی تعداد چانک‌ها

- تعداد چانک‌های تولید شده برای هر فایل صوتی در صورت فعال بودن گزینه `verbose` چاپ می‌شود:
    ```python
    if verbose:
        print(f" => Splitting into {num_chunks} chunks.")
    ```

### نتیجه بخش دوم

- فایل‌های صوتی براساس مدت زمان چانک و مقدار همپوشانی به بخش‌های کوچکتر تقسیم شدند.
- تمامی چانک‌ها دارای طول یکنواخت هستند و آماده مراحل بعدی پردازش، مانند افزایش داده و استخراج ویژگی‌های MFCC، می‌باشند.
- در بخش سوم، به فرآیند افزایش داده (Data Augmentation) و استخراج ویژگی‌ها پرداخته خواهد شد.

# بخش سوم: افزایش داده (Data Augmentation) و استخراج ویژگی‌های MFCC

این بخش ترکیبی از دو فرآیند کلیدی را شامل می‌شود: **افزایش داده** و **استخراج ویژگی‌های MFCC پیشرفته**. ابتدا داده‌های صوتی با استفاده از تکنیک‌های افزایش داده غنی‌تر می‌شوند. سپس ویژگی‌های MFCC و مشتقات آن استخراج می‌شوند تا داده‌ها برای مدل یادگیری عمیق آماده شوند.

---

## افزایش داده (Data Augmentation)

- **هدف**: افزایش تنوع داده‌ها برای جلوگیری از **اورفیتینگ** و بهبود تعمیم‌پذیری مدل.
- **تکنیک‌های مورد استفاده**:
  1. **افزودن نویز تصادفی**:
     - نویز با شدت کم به سیگنال صوتی اضافه می‌شود.
  2. **تغییر گام (Pitch Shift)**:
     - فرکانس گام سیگنال صوتی تغییر می‌کند.
  3. **تغییر سرعت پخش (Time Stretch)**:
     - سرعت سیگنال صوتی افزایش یا کاهش می‌یابد.
- **اعمال تغییرات تصادفی**:
  - هر تکنیک با احتمال مشخص (مانند 40%) روی داده اعمال می‌شود:
    ```python
    chunk = augment_audio(chunk, sr_loaded, p=0.4)
    ```
- **مزایا**:
  - افزایش تنوع داده‌ها.
  - بهبود مقاومت مدل در برابر تغییرات صوتی مانند نویز محیطی.

---

## استخراج ویژگی‌های MFCC پیشرفته

### 1. **محاسبه MFCC**
- **MFCC اصلی**:
  - این ویژگی‌ها انرژی فرکانس‌های مختلف سیگنال را در مقیاس فرکانسی مل نشان می‌دهند:
    ```python
    mfccs = librosa.feature.mfcc(y=chunk, sr=sr_loaded, n_mfcc=N_MFCC)
    ```

---

### 2. **نرمال‌سازی CMVN**
- **تعریف**:
  - نرمال‌سازی میانگین و واریانس MFCC برای حذف تأثیر شدت صدا.
- **فرمول محاسبه**:
    ```python
    mfccs = cmvn(mfccs)
    ```

---

### 3. **محاسبه دلتا و دلتا-دلتا**
- **تعریف**:
  - مشتقات مرتبه اول (دلتا) و مرتبه دوم (دلتا-دلتا) MFCC به عنوان ویژگی‌های زمانی استخراج می‌شوند.
- **محاسبه دلتا و دلتا-دلتا**:
    ```python
    mfcc_delta = librosa.feature.delta(mfccs, order=1)
    mfcc_delta2 = librosa.feature.delta(mfccs, order=2)
    ```

---

### 4. **ادغام ویژگی‌ها**
- **تعریف**:
  - ویژگی‌های MFCC، دلتا، و دلتا-دلتا ترکیب می‌شوند تا یک آرایه جامع برای هر چانک صوتی ایجاد شود:
    ```python
    mfcc_cat = np.concatenate([mfccs, mfcc_delta, mfcc_delta2], axis=0)
    ```

---

## تغییر اندازه و نرمال‌سازی نهایی

- **تغییر اندازه (Resize)**:
  - داده‌های ترکیبی به اندازه استاندارد برای مدل یادگیری عمیق تغییر داده می‌شوند:
    ```python
    mfcc_resized = resize(mfcc_cat, (*TARGET_SHAPE, 1))
    ```

- **نرمال‌سازی نهایی (اختیاری)**:
  - مقادیر ویژگی‌ها به بازه `[0, 1]` مقیاس‌بندی می‌شوند:
    ```python
    if normalize:
        _min, _max = mfcc_resized.min(), mfcc_resized.max()
        mfcc_resized = (mfcc_resized - _min) / (_max - _min)
    ```

---

## ذخیره داده‌ها و برچسب‌ها

- **اضافه کردن داده‌ها و برچسب‌ها**:
  - ویژگی‌های استخراج‌شده به لیست `data` اضافه می‌شوند.
  - برچسب کلاس مربوطه به لیست `labels` اضافه می‌شود:
    ```python
    data.append(mfcc_resized)
    labels.append(class_idx)
    ```

---

### نتیجه بخش سوم

- **افزایش داده**:
  - داده‌های صوتی با تکنیک‌های افزایش داده، متنوع‌تر شدند.
- **استخراج ویژگی‌های MFCC**:
  - ویژگی‌های MFCC و مشتقات آن استخراج و به شکل یکنواخت آماده شدند.
- **آماده‌سازی داده‌ها**:
  - داده‌های پردازش‌شده و برچسب‌های مربوطه آماده استفاده در مدل یادگیری عمیق شدند.

</div>


````python
def load_and_preprocess_data(
    data_dir,
    classes,
    chunk_duration=4.0,
    overlap_duration=2.0,
    sr=None,
    augmentation=False,
    trim_silence=False,
    top_db=20,
    verbose=True
):
    data = []
    labels = []

    if verbose:
        print("[INFO] Loading dataset from:", data_dir)
        print(f"[INFO] Classes: {classes}")
        print(f"[INFO] chunk_duration={chunk_duration}s, overlap={overlap_duration}s")
        print(f"[INFO] augmentation={augmentation}, trim_silence={trim_silence}")

    for class_idx, class_name in enumerate(classes):
        class_folder = os.path.join(data_dir, class_name)
        if not os.path.isdir(class_folder):
            print(f"[WARNING] Skip {class_folder}: not a directory.")
            continue

        if verbose:
            print(f"\n[INFO] Processing class: {class_name}")

        for file_name in os.listdir(class_folder):
            if not file_name.lower().endswith(".wav"):
                continue

            file_path = os.path.join(class_folder, file_name)
            if verbose:
                print(f"   -> Loading {file_name} ... ", end='')

            try:
                audio_data, sr_loaded = librosa.load(file_path, sr=sr)
            except Exception as e:
                print(f"\n[ERROR] Could not load file: {file_name}, error: {e}")
                continue

            if verbose:
                print("OK.", end=" ")

            # حذف سکوت (اختیاری)
            if trim_silence:
                trimmed_data, _ = librosa.effects.trim(audio_data, top_db=top_db)
                if len(trimmed_data) > 0:
                    audio_data = trimmed_data

            chunk_samples = int(chunk_duration * sr_loaded)
            overlap_samples = int(overlap_duration * sr_loaded)

            # بررسی کوتاه بودن فایل
            if len(audio_data) < chunk_samples:
                print(f"[WARNING] {file_name} is shorter than {chunk_samples} samples. Skipped.")
                continue

            num_chunks = int(np.ceil((len(audio_data) - chunk_samples) / (chunk_samples - overlap_samples))) + 1

            if verbose:
                print(f" => Splitting into {num_chunks} chunks.")

            for i_chunk in range(num_chunks):
                start = i_chunk * (chunk_samples - overlap_samples)
                end = start + chunk_samples
                chunk = audio_data[start:end]

                # پدینگ چانک‌های کوتاه‌تر
                if len(chunk) < chunk_samples:
                    pad_size = chunk_samples - len(chunk)
                    chunk = np.pad(chunk, (0, pad_size), mode='constant')

                # افزایش داده (اختیاری)
                if augmentation:
                    chunk = augment_audio(chunk, sr_loaded, p=0.4)

                # استخراج ویژگی PLP
                feature_data = extract_plp_features(
                    chunk,
                    sr=sr_loaded,
                    n_plp=N_PLP,
                    apply_cmvn=True,
                    resize_shape=TARGET_SHAPE,
                    normalize=False
                )

                data.append(feature_data)
                labels.append(class_idx)

    data = np.array(data, dtype=np.float32)
    labels = np.array(labels, dtype=np.int32)

    if verbose:
        print("\n[INFO] Finished processing.")
        print(f"       Data shape : {data.shape}")
        print(f"       Labels shape : {labels.shape}")
        if data.size > 0:
            print(f"       Data range  : {data.min()} to {data.max()}")

    return data, labels
````

````python


# =================================================
# 5. ساخت مدل با معماری دقیقا همان که خواستید
# =================================================
def build_model(input_shape, num_classes):
    """
    عین معماری اصلی داده شده:
    2 * (Conv 32) -> MaxPool
    2 * (Conv 64) -> MaxPool
    2 * (Conv 128) -> MaxPool
    Dropout(0.3)
    2 * (Conv 256) -> MaxPool
    2 * (Conv 512) -> MaxPool
    Dropout(0.3)
    Flatten
    Dense(1200, relu)
    Dropout(0.45)
    Dense(num_classes, softmax)
    """
    from tensorflow.keras.models import Sequential

    model = Sequential()
    # همان معماری
    model.add(Conv2D(filters=32, kernel_size=3, padding='same', activation='relu',
                     input_shape=input_shape))
    model.add(Conv2D(filters=32, kernel_size=3, activation='relu'))
    model.add(MaxPool2D(pool_size=2, strides=2))

    model.add(Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'))
    model.add(Conv2D(filters=64, kernel_size=3, activation='relu'))
    model.add(MaxPool2D(pool_size=2, strides=2))

    model.add(Conv2D(filters=128, kernel_size=3, padding='same', activation='relu'))
    model.add(Conv2D(filters=128, kernel_size=3, activation='relu'))
    model.add(MaxPool2D(pool_size=2, strides=2))

    model.add(Dropout(0.3))

    model.add(Conv2D(filters=256, kernel_size=3, padding='same', activation='relu'))
    model.add(Conv2D(filters=256, kernel_size=3, activation='relu'))
    model.add(MaxPool2D(pool_size=2, strides=2))

    model.add(Conv2D(filters=512, kernel_size=3, padding='same', activation='relu'))
    model.add(Conv2D(filters=512, kernel_size=3, activation='relu'))
    model.add(MaxPool2D(pool_size=2, strides=2))

    model.add(Dropout(0.3))

    model.add(Flatten())
    model.add(Dense(units=1200, activation='relu'))
    model.add(Dropout(0.45))

    # خروجی
    model.add(Dense(units=num_classes, activation='softmax'))

    return model
````

````python
# =========================
# 6. قسمت اصلی اجرا (main)
# =========================
def main():
    data, labels = load_and_preprocess_data(
        data_dir=DATA_DIR,
        classes=CLASSES,
        chunk_duration=CHUNK_DURATION,
        overlap_duration=OVERLAP_DURATION,
        sr=None,
        augmentation=False,
        trim_silence=True,
        top_db=20,
        verbose=True
    )


    # تبدیل لیبل‌ها به one-hot
    labels = to_categorical(labels, num_classes=len(CLASSES))
    def show_three_samples(chunks, sr):
        # بررسی می‌کنیم که حداقل 3 چانک وجود داشته باشد
        if len(chunks) < 3:
            print("[WARNING] Not enough chunks to show (need at least 3).")
            return

        # انتخاب تصادفی 3 چانک
        selected_indices = random.sample(range(len(chunks)), 3)

        for i, idx in enumerate(selected_indices):
            print(f"\n[INFO] Showing MFCC for chunk index = {idx}")
            chunk = chunks[idx]
            show_mfcc_before_after(audio_data=chunk, sr=sr, n_mfcc=20, target_shape=(150, 150))
    # تقسیم آموزش/آزمون
    X_train, X_test, y_train, y_test = train_test_split(
        data, labels,
        test_size=0.2,
        random_state=42
    )

    print("\n[INFO] Split Data:")
    print(f"   X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
    print(f"   X_test shape : {X_test.shape},  y_test shape : {y_test.shape}")

    # --- بخش ب: ساخت مدل با معماری کاملا یکسان ---
    model = build_model(input_shape=X_train.shape[1:], num_classes=len(CLASSES))
    model.summary()

    # کامپایل
    optimizer = Adam(learning_rate=1e-4)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])


    # --- تعریف کالبک‌های پیشرفته ---
    from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, CSVLogger, Callback

    # کالبک توقف زودهنگام
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True,
        verbose=1
    )

    # کالبک کاهش نرخ یادگیری در صورت عدم پیشرفت
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-6,
        verbose=1
    )

    # کالبک ذخیره بهترین مدل
    checkpoint = ModelCheckpoint(
        filepath='best_model_with_mfcc.keras',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    )

    # کالبک لاگ CSV
    csv_logger = CSVLogger(
        filename='training_log_with_mfcc.csv',
        append=True
    )

    # کالبک سفارشی (مثال: اگر val_accuracy > 0.90 شد، متوقف شو)
    class CustomCallback(Callback):
        def on_epoch_end(self, epoch, logs=None):
            logs = logs or {}
            val_acc = logs.get('val_accuracy')
            if val_acc and val_acc > 0.90:
                print(f"\n[INFO] Validation Accuracy reached {val_acc:.2f}. Stopping training...")
                self.model.stop_training = True

    custom_callback = CustomCallback()

    # لیست کالبک‌ها
    callbacks_list = [early_stopping, reduce_lr, checkpoint, csv_logger, custom_callback]

    # --- بخش ج: آموزش ---
    history = model.fit(
        X_train, y_train,
        epochs=50,
        batch_size=32,
        validation_data=(X_test, y_test),
        callbacks=callbacks_list,  # اضافه کردن کالبک‌ها
        verbose=1
    )

    # ذخیره مدل نهایی (آخرین وزن‌ها)
    model.save("Trained_model_with mfcc.keras")

    # ذخیره تاریخچه آموزش (اختیاری)
    with open("training_history _with_mfcc.json", "w") as f:
        json.dump(history.history, f)

    # --- ارزیابی ---
    train_loss, train_acc = model.evaluate(X_train, y_train, verbose=0)
    test_loss, test_acc   = model.evaluate(X_test, y_test, verbose=0)

    print(f"\n[RESULT] Training Accuracy: {train_acc:.4f}")
    print(f"[RESULT] Testing Accuracy : {test_acc:.4f}")

    # --- رسم نمودار Loss و Accuracy ---
    plot_learning_curves(history)

    # --- گزارش و ماتریس درهم‌ریختگی ---
    y_pred = model.predict(X_test, verbose=0)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_test, axis=1)

    print("\n[CLASSIFICATION REPORT]")
    print(classification_report(y_true_classes, y_pred_classes, target_names=CLASSES))

    cm = confusion_matrix(y_true_classes, y_pred_classes)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='d',
                xticklabels=CLASSES, yticklabels=CLASSES)
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted Class")
    plt.ylabel("Actual Class")
    plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import librosa
from skimage.transform import resize

def show_mfcc_before_after(audio_data, sr, n_mfcc=20, target_shape=(150, 150)):
    # 1) محاسبه MFCC اصلی
    mfccs = librosa.feature.mfcc(y=audio_data, sr=sr, n_mfcc=n_mfcc)

    # 2) محاسبه دلتا و دلتا-دلتا
    mfcc_delta  = librosa.feature.delta(mfccs, order=1)
    mfcc_delta2 = librosa.feature.delta(mfccs, order=2)

    # 3) اتصال در محور فرکانس: (3*n_mfcc, time_frames)
    mfcc_cat = np.concatenate([mfccs, mfcc_delta, mfcc_delta2], axis=0)

    # 4) یک کپی برای قبل از Resize
    mfcc_original = mfcc_cat.copy()

    # 5) تبدیل به شکل (..., 1)
    mfcc_cat = np.expand_dims(mfcc_cat, axis=-1)  # (3*n_mfcc, time_frames, 1)

    # 6) Resize برای بعد از
    mfcc_resized = resize(mfcc_cat, (*target_shape, 1))

    # حالا دو تصویر را در کنار هم رسم می‌کنیم
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    # -- تصویر قبل از Resize --
    # توجه: برای نشان دادن در imshow باید بعد سوم (کانال) را حذف کنیم یا همان mfcc_original استفاده کنیم
    im0 = axes[0].imshow(mfcc_original, aspect='auto', origin='lower', cmap='viridis')
    axes[0].set_title(f"Before Resize\nshape={mfcc_original.shape}")
    fig.colorbar(im0, ax=axes[0])

    # -- تصویر بعد از Resize --
    im1 = axes[1].imshow(mfcc_resized[:, :, 0], aspect='auto', origin='lower', cmap='viridis')
    axes[1].set_title(f"After Resize\nshape={mfcc_resized.shape}")
    fig.colorbar(im1, ax=axes[1])

    plt.tight_layout()
    plt.show()
````

````python

def plot_learning_curves(history):
    """
    رسم نمودارهای Loss و Accuracy
    """
    epochs = range(1, len(history.history['loss']) + 1)

    # نمودار Loss
    plt.figure(figsize=(7, 5))
    plt.plot(epochs, history.history['loss'], label='Train Loss', color='red')
    plt.plot(epochs, history.history['val_loss'], label='Val Loss', color='blue')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Loss over Epochs')
    plt.legend()
    plt.show()

    # نمودار Accuracy
    plt.figure(figsize=(7, 5))
    plt.plot(epochs, history.history['accuracy'], label='Train Accuracy', color='red')
    plt.plot(epochs, history.history['val_accuracy'], label='Val Accuracy', color='blue')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Accuracy over Epochs')
    plt.legend()
    plt.show()


# ====================================
# اجرای اصلی در صورت اجرای مستقیم فایل
# ====================================
if __name__ == "__main__":
    main()
````

````python
def load_and_preprocess_data_with_visualization(
    data_dir,
    classes,
    chunk_duration=4.0,
    overlap_duration=2.0,
    sr=None,
    augmentation=False,
    trim_silence=False,
    top_db=20,
    verbose=True,
    visualize_chunks=3  # تعداد چانک‌هایی که قرار است بصری‌سازی شوند
):
    """
    این تابع تمام داده‌ها را پردازش کرده و ویژگی‌های MFCC
    قبل و بعد از Resize برای چند چانک مشخص بصری‌سازی می‌کند.
    """
    data = []
    labels = []

    if verbose:
        print("[INFO] Loading dataset from:", data_dir)
        print(f"[INFO] Classes: {classes}")
        print(f"[INFO] chunk_duration={chunk_duration}s, overlap={overlap_duration}s")
        print(f"[INFO] augmentation={augmentation}, trim_silence={trim_silence}")

    # شمارنده برای نمایش چانک‌ها
    visualized_count = 0

    for class_idx, class_name in enumerate(classes):
        class_folder = os.path.join(data_dir, class_name)
        if not os.path.isdir(class_folder):
            print(f"[WARNING] Skip {class_folder}: not a directory.")
            continue

        if verbose:
            print(f"\n[INFO] Processing class: {class_name}")

        for file_name in os.listdir(class_folder):
            if not file_name.lower().endswith(".wav"):
                continue

            file_path = os.path.join(class_folder, file_name)
            if verbose:
                print(f"   -> Loading {file_name} ... ", end='')

            try:
                audio_data, sr_loaded = librosa.load(file_path, sr=sr)
            except Exception as e:
                print(f"\n[ERROR] Could not load file: {file_name}, error: {e}")
                continue

            if verbose:
                print("OK.", end=" ")

            # حذف سکوت (اختیاری)
            if trim_silence:
                trimmed_data, _ = librosa.effects.trim(audio_data, top_db=top_db)
                if len(trimmed_data) > 0:
                    audio_data = trimmed_data

            chunk_samples = int(chunk_duration * sr_loaded)
            overlap_samples = int(overlap_duration * sr_loaded)

            # بررسی کوتاه بودن فایل
            if len(audio_data) < chunk_samples:
                print(f"[WARNING] {file_name} is shorter than {chunk_samples} samples. Skipped.")
                continue

            num_chunks = int(np.ceil((len(audio_data) - chunk_samples) / (chunk_samples - overlap_samples))) + 1

            if verbose:
                print(f" => Splitting into {num_chunks} chunks.")

            for i_chunk in range(num_chunks):
                start = i_chunk * (chunk_samples - overlap_samples)
                end = start + chunk_samples
                chunk = audio_data[start:end]

                # پدینگ چانک‌های کوتاه‌تر
                if len(chunk) < chunk_samples:
                    pad_size = chunk_samples - len(chunk)
                    chunk = np.pad(chunk, (0, pad_size), mode='constant')

                # افزایش داده (اختیاری)
                if augmentation:
                    chunk = augment_audio(chunk, sr_loaded, p=0.4)

                # استخراج ویژگی MFCC
                feature_data = extract_mfcc_features(
                    chunk,
                    sr=sr_loaded,
                    n_mfcc=N_MFCC,
                    apply_cmvn=True,
                    resize_shape=TARGET_SHAPE,
                    normalize=False
                )

                # ذخیره داده‌ها
                data.append(feature_data)
                labels.append(class_idx)

                # بصری‌سازی MFCC قبل و بعد از Resize
                if visualized_count < visualize_chunks:
                    print(f"\n[INFO] Visualizing chunk {visualized_count + 1}/{visualize_chunks} ...")
                    show_mfcc_before_after(chunk, sr=sr_loaded, n_mfcc=N_MFCC, target_shape=TARGET_SHAPE)
                    visualized_count += 1

    data = np.array(data, dtype=np.float32)
    labels = np.array(labels, dtype=np.int32)

    if verbose:
        print("\n[INFO] Finished processing.")
        print(f"       Data shape : {data.shape}")
        print(f"       Labels shape : {labels.shape}")
        if data.size > 0:
            print(f"       Data range  : {data.min()} to {data.max()}")

    return data, labels

data, labels = load_and_preprocess_data_with_visualization(
    data_dir=DATA_DIR,
    classes=CLASSES,
    chunk_duration=4.0,
    overlap_duration=2.0,
    sr=None,  # None: استفاده از نرخ نمونه‌برداری اصلی
    augmentation=True,  # افزایش داده فعال
    trim_silence=True,  # حذف سکوت فعال
    top_db=20,
    verbose=True,
    visualize_chunks=3  # تعداد چانک‌هایی که نمایش داده می‌شود
)
````
